## Select 50 images per dx column to create 350 images for training

```python
# Function to select 50 random images per diagnosis type
def select_50_images_per_diagnosis(metadata, combined_folder):
    # Get unique diagnosis types
    diagnosis_types = metadata['dx'].unique()

    # Initialize an empty list to store selected cases
    selected_cases = []

    # Create a folder for the selected images
    output_folder = '/content/train_50_images_per_dx'
    os.makedirs(output_folder, exist_ok=True)

    # Iterate through each unique diagnosis type
    for dx_type in diagnosis_types:
        # Filter metadata for a specific diagnosis type
        df = metadata[metadata['dx'] == dx_type]

        # Check if there are at least 50 images available for this diagnosis type
        if len(df) >= 50:
            # Select 50 random images for the diagnosis type
            random_images = df.sample(n=50, random_state=1)

            # Move the selected images to the folder
            for _, row in random_images.iterrows():
                source_path = os.path.join(combined_folder, row['image_id_with_extension'])
                destination_path = os.path.join(output_folder, row['image_id_with_extension'])
                shutil.copy(source_path, destination_path)

            # Store selected metadata
            selected_cases.append(random_images)

    # Create output metadata file
    selected_metadata = pd.concat(selected_cases)
    selected_metadata_path = '/content/train_50_images_per_dx/selected_metadata_random_50_case_per_dx.csv'
    selected_metadata.to_csv(selected_metadata_path, index=False)

    return selected_metadata['image_id_with_extension'].tolist()

# Usage
removed_images_50 = select_50_images_per_diagnosis(metadata, '/content/combined_folder')
```

## Select 1 image per dx column for later testing

```python
# Function to select one random image per diagnosis type
def select_random_image_per_diagnosis(metadata, combined_folder):
    # Get unique diagnosis types
    diagnosis_types = metadata['dx'].unique()

    # Initialize an empty list to store selected cases
    selected_cases = []

    # Create a folder for the selected images
    output_folder = '/content/test_1_image_per_dx'
    os.makedirs(output_folder, exist_ok=True)

    # Iterate through each unique diagnosis type
    for dx_type in diagnosis_types:
        # Filter metadata for a specific diagnosis type
        df = metadata[metadata['dx'] == dx_type]

        # Check if there are images available for this diagnosis type
        if not df.empty:
            # Select a random image for the diagnosis type
            random_image = df.sample(n=1, random_state=1)

            # Move the selected image to the folder
            source_path = os.path.join(combined_folder, random_image['image_id_with_extension'].iloc[0])
            destination_path = os.path.join(output_folder, random_image['image_id_with_extension'].iloc[0])
            shutil.copy(source_path, destination_path)

            # Store selected metadata
            selected_cases.append(random_image)

    # Create output metadata file
    selected_metadata = pd.concat(selected_cases)
    selected_metadata_path = '/content/test_1_image_per_dx/selected_metadata_random_case_per_dx.csv'
    selected_metadata.to_csv(selected_metadata_path, index=False)

    return selected_metadata['image_id_with_extension'].tolist()

# Usage
removed_images = select_random_image_per_diagnosis(metadata, '/content/combined_folder')
```

## 1. Model Building and Training:

```python
# function to create train adn validation generators
def create_data_generators(subset_size, json_filename):
    metadata_subset = metadata.sample(n=subset_size)
    image_folder_path = '/content/combined_folder'
    img_height, img_width = 28, 28
    batch_size = 128  # Adjusted batch size

    train_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.2,
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True
    )

    train_generator = train_datagen.flow_from_dataframe(
        dataframe=metadata_subset,
        directory=image_folder_path,
        x_col='image_id_with_extension',
        y_col='diagnosis_no',  # Use 'diagnosis_no' here as the target variable
        target_size=(img_height, img_width),
        batch_size=batch_size,
        class_mode='categorical',
        subset='training'
    )

    validation_generator = train_datagen.flow_from_dataframe(
        dataframe=metadata_subset,
        directory=image_folder_path,
        x_col='image_id_with_extension',
        y_col='diagnosis_no',  # Use 'diagnosis_no' here as the target variable
        target_size=(img_height, img_width),
        batch_size=batch_size,
        class_mode='categorical',
        subset='validation'
    )

    # Extract the parameters used for augmentation
    augmentation_parameters = {
        'rescale': train_datagen.rescale,
        'rotation_range': train_datagen.rotation_range,
        'width_shift_range': train_datagen.width_shift_range,
        'height_shift_range': train_datagen.height_shift_range,
        'shear_range': train_datagen.shear_range,
        'zoom_range': train_datagen.zoom_range,
        'horizontal_flip': train_datagen.horizontal_flip
    }

    # Save the augmentation parameters to a JSON file with the specified filename
    with open(json_filename, 'w') as file:
        json.dump(augmentation_parameters, file)

    return train_generator, validation_generator
```

```python
# function to define the model architecture
def create_model(img_height, img_width, train_generator, validation_generator, epochs=5, num_classes=7):
    model = Sequential()

    # Add Convolutional layers
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)))
    model.add(MaxPooling2D((2, 2)))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    # Additional Convolutional layers for increased complexity
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))

    model.add(Flatten())

    # Adjusted Dense layers for compatibility
    model.add(Dense(256, activation='relu'))  # Increased units in the Dense layer
    model.add(Dropout(0.5))
    model.add(Dense(128, activation='relu'))  # Additional Dense layer
    model.add(Dropout(0.5))

    model.add(Dense(num_classes, activation='softmax'))

    opt = Adam(learning_rate=0.0001)
    model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

    early_stopping = EarlyStopping(patience=3, restore_best_weights=True)

    history = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=validation_generator,
        callbacks=[early_stopping]
    )

    return model, history
```

```python
# Running create_data_generators() and create_model() functions to create the model and check history

# Define img_height, img_width, num_classes, train_generator, validation_generator, epochs
img_height, img_width = 28, 28
num_classes = 7
epochs = 10
batch_size = 128
subset_size = 10000  # Adjust this value as needed
json_filename = '/content/Model3_28_10000_augmentation_params.json'
metadata['diagnosis_no'] = metadata['diagnosis_no'].astype(str)   # diagnosis_no has to be string type due to class_mode=categorical

# Run the function to create generators
train_gen, validation_gen = create_data_generators(subset_size=10000, json_filename=json_filename)

# Create and train the model
model_28_10000, history_28_10000 = create_model(img_height, img_width, train_gen, validation_gen, epochs=epochs, num_classes=num_classes)
```

## Visualize Training and Validation History

```python
# define function to visualize the result
def plot_history(history):
    # Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
```

```python
# Visualize the training history
plot_history(history_28_10000)
```
